```python
annote = []
images = []
for i in range(1,1601):
    path = "/content/dataset/"+str(i)+".jpg"
    img = cv2.imread(path)
    img1 = Image.open(path)
    width, height = img1.size
    gr_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    r, bw_img = cv2.threshold(gr_img, 127, 255, cv2.THRESH_BINARY)
    fortran_ground_truth_binary_mak = np.asfortranarray(bw_img)
    encoded_ground_truth = mask.encode(fortran_ground_truth_binary_mask)
    ground_truth_area = mask.area(encoded_ground_truth)
    ground_truth_bounding_box = mask.toBbox(encoded_ground_truth)
    contours = measure.find_contours(bw_img, 0.5)
    annotation = {
        "segmentation":[],
        "area":ground_truth_area.tolist(),
        "iscrowd":0,
        "image_id":i,
        "bbox":ground_truth_bounding_box.tolist(),
        "category_id":1,
        "id":i-1
    }
    info = {
        "id":i,
        "width":width,
        "height":height,
        "file_name":str(i)+.".jpg"
    }
    for contour in contours:
        contour = np.flip(contour, axis = 1)
        segmentation = contour.ravel().tolist()
        annotation["segmentation"].append(segmentation)
    annote.append(annotation)
    images.append(info)
b = json.dumps(images)
a = json.dumps(annote)
f1 = open("/content/annotations-3.json","w")
f1.writelines(b)
f1.writelines(a)
f1.close()
```
